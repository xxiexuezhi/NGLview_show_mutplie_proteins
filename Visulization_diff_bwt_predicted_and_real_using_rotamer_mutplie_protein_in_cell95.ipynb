```python
def get_mainchain_ag(file_address):
    #take a pdb file return the main chain atom group as N, CA,C 
    u2 = mda.Universe(file_address, topology_format='PDB')
    mc_protein = u2.select_atoms("name N CA C")
    return mc_protein


def get_lev1_ag(file_address):
    u2 = mda.Universe(file_address, topology_format='PDB')
    protein = u2.select_atoms("all")
    index = []
    for atom in protein:
        if atom.name in dict_lev1_2[atom.resname.upper()]:
            index.append(atom.index)
    return protein[index]


def get_all_sidechain(file_address):
    u2 = mda.Universe(file_address, topology_format='PDB')
    protein = u2.select_atoms("all and not backbone")
    return protein
    
```

```python
import rmsd
dict_mc_lev1_2 = {'ARG': ['C','N','NH1', 'CZ', 'NE'],
 'THR': ['C','N','CB', 'CG2', 'CA'],
 'ASP': ['C','N','CG', 'OD1', 'CB'],
 'TYR': ['C','N','CZ', 'OH', 'CB', 'CG', 'CD2'],
 'PHE': ['C','N','CB', 'CG', 'CD2'],
 'GLU': ['C','N','CD', 'OE1', 'CG'],
 'ILE': ['C','N','CG1', 'CD1', 'CB'],
 'LEU': ['C','N','CG', 'CD2', 'CB'],
 'LYS': ['C','N','CE', 'NZ', 'CD'],
 'CYS': ['C','N','CB', 'SG', 'CA'],
 'SER': ['C','N','CB', 'OG', 'CA'],
 'ASN': ['C','N','CG', 'OD1', 'CB'],
 'MET': ['C','N','SD', 'CE', 'CG'],
 'GLN': ['C','N','CD', 'NE2', 'CG'],
 'PRO': ['C','N','CB', 'CG', 'CA'],
 'HIS': ['C','N','CB', 'CG', 'ND1'],
 'TRP': ['C','N','CB', 'CG', 'CD2'],
 'VAL': ['C','N','CA', 'CB'],
 'GLY': ['C','N','CA'],
 'ALA': ['C','N','CA', 'CB']}


def get_atoms_mc_lev1_2(pdb_file_address):
    lst_coord = []
    p = PDBParser()
    structure = p.get_structure("X", pdb_file_address)
    ppb = PPBuilder()
    for res in ppb.build_peptides(structure)[0]:
        res3_name = res.get_resname().upper()
        lev1_atom_name_lst = dict_mc_lev1_2[res3_name]
        atom_coord_per_res = []
        for lev1_atom_name in lev1_atom_name_lst:
            #print(lev1_atom_name)
            lev1_coord = res[lev1_atom_name].get_coord()
            atom_coord_per_res.append(lev1_coord)
        lst_coord.append(atom_coord_per_res)
    return np.array(lst_coord)


def change_format(lst_np):
    lst_changed=[]
    for lst_element in lst_np:
        for arr in lst_element:
            lst_changed.append(arr)
    return np.array(lst_changed)
        


rmsd_lst = []
for i in range(4,10):
    model_file = filtered_lst[j]
    rotamer_file = "./rotamer_g_data/g_rotamer_q_test_"+str(i)+".pdb"
    decoded_lev12_0=get_atoms_mc_lev1_2(rotamer_file)
    native_lev12_0 = get_atoms_mc_lev1_2(filtered_lst[i])
    value = rmsd.rmsd(change_format(decoded_lev12_0),change_format(native_lev12_0))
    rmsd_lst.append(value)
```
